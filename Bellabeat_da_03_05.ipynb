## Data cleaning

## Covert Date column from string to date

Before making any analysis the data set requires to change the data type of the date column (its a text/string) into a date format, if this step is omitted any query which uses the date column won't work.
The first step is adding a new column with the right data type.
```js
ALTER TABLE daily_activity_03_05
ADD COLUMN activitydate;
```
Then we fill this new using the following query:
```js
UPDATE daily_activity_03_05
SET activitydate = TO_DATE(date, 'YYYY/MM/DD');
```

## Duplicates

````python
sql4= """SELECT Id, activitydate, TotalSteps, COUNT(*)
FROM public.daily_activity_03_05
GROUP BY Id, activitydate, TotalSteps
HAVING COUNT(*) > 1;"""
df4 = sqlio.read_sql_query(sql4 ,conn2)
display(df4)
````

- Our query showed any record, it means that our data set doesn't have any duplicates.

## Remove columns
When we updated the date column in the first step of the data cleaning process we ended up with two date columns, for this reason is a must to remove the original Date column to finish the data cleaning process, futhermore the column named trackerdistance is equals to the totaldistance column so we is a must to remove that column as well.
The following query removed both date columns:  
```js

ALTER TABLE daily_activity_03_05
DROP COLUMN trackerdistance;

ALTER TABLE daily_activity_03_05
DROP COLUMN date;

```
